```python
tr_customer = pd.read_csv('/Users/sofiede.cnudde/Downloads/GNN_returns_prediction_dataset_v2/customer_nodes_training.csv')
te_customer = pd.read_csv('/Users/sofiede.cnudde/Downloads/GNN_returns_prediction_dataset_v2/customer_nodes_testing.csv')

tr_product = pd.read_csv('/Users/sofiede.cnudde/Downloads/GNN_returns_prediction_dataset_v2/product_nodes_training.csv')
te_product = pd.read_csv('/Users/sofiede.cnudde/Downloads/GNN_returns_prediction_dataset_v2/product_nodes_testing.csv')

tr_event = pd.read_csv('/Users/sofiede.cnudde/Downloads/GNN_returns_prediction_dataset_v2/event_table_training.csv')
te_event = pd.read_csv('/Users/sofiede.cnudde/Downloads/GNN_returns_prediction_dataset_v2/event_table_testing.csv')
```

```python
tr_customer.insert(0, 'customer_id', range(0, 0 + len(tr_customer)))
```

```python
tr_customer
```

```python
tr_product.insert(0, 'product_id', range(649119, 649119 + len(tr_product)))
```

```python
tr_product
```

```python
edges = tr_event.merge(tr_customer[['hash(customerId)', 'customer_id']], on='hash(customerId)')
edges = edges.merge(tr_product[['variantID', 'product_id']], on='variantID')
edges
```

```python
customer_nodes = tr_customer.set_index('customer_id')
```

```python
product_nodes = tr_product.set_index('product_id')
```

```python
customer_nodes = customer_nodes[['yearOfBirth', 'isMale']]
customer_nodes
```

```python
product_nodes = product_nodes[['productID', 'avgGbpPrice']]
product_nodes
```

```python
df_returned = edges[edges.isReturned == 1]
```

```python
df_returned = df_returned[['customer_id', 'product_id']]
```

```python
df_returned
```

```python
returns = StellarGraph({"customers": customer_nodes, "products": product_nodes},
                       df_returned, source_column="customer_id", target_column="product_id")
print(returns.info())
```

```python
df_events = edges[['customer_id', 'product_id', 'isReturned']]
```

```python
purchases_returns = StellarGraph({"customers": customer_nodes, "products": product_nodes},
                                 df_events, source_column="customer_id", target_column="product_id",
                                 edge_type_column="isReturned")
print(purchases_returns.info())
```
